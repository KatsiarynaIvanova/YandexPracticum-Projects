## Предобработка данных

```python
# Переименовываем столбце cityCenters_neares
data = data.rename(columns={'cityCenters_nearest':'city_centers_nearest'})
```

```python
# NaN в столбце balcony может означать отсутсвие балкона. Заменем такие пропуски 0.
data['balcony'] = data['balcony'].fillna(0)
data.isna().sum()
```

```python
# Вероятнее всего, что значения NaN в столбце is_apartment соответствуют значению False. Заменем NaN на False.
data['is_apartment'] = data['is_apartment'].fillna(False)
data['is_apartment'].unique()
```

Пропуски в столбце floors_total можно заменить значениями из столбца floor. Возможно, человек не указал общую этажность дома, так как живет на последнем этаже.

```python
# Заменяем пропуски в floors_total значениями из floor
data['floors_total'] = data['floors_total'].fillna(data['floor'])
data['floors_total'].isna().sum()
```

```python
# Смотрим, есть ли дубликаты
data.duplicated().sum()
```

```python
# Приводим столбец locality_name к нижнему регистру и смотри количество дубликатов снова
data['locality_name'] = data['locality_name'].str.lower()
data.duplicated().sum()
```

#### Заполнение пропусков в столбце ceiling_height

```python
# корреляция между высотой потолков и количеством этажей. Вывод: отсутствует.
data['floors_total'].corr(data.ceiling_height)
```

```python
# смотрим диаграмму размаха для количества этажей и высоты потолка
data.boxplot(
    by='floors_total', 
    column='ceiling_height',
    figsize=(15, 8)
)
```

Обнаружены выбросы, которые не соответсвуют реальности, например, высота потолка 100 м. Избавимся от выбросов. 

```python
# Отфильтруем data по ceiling_height превышающим 10 м.
data[data.ceiling_height >= 10][['floors_total', 'ceiling_height']].head(75)
```

Значения высоты пололков превышающих 20 метров делим на 10. Предплогаем ошибку в записи данных, поскольку тип столбца float могла произойти ошибка при записи данных.

```python
# Делим значения на 10 и строим диаграмму размаха
data.loc[data.ceiling_height >= 20, 'ceiling_height'] = data.ceiling_height / 10
data.boxplot(
    by='floors_total', 
    column='ceiling_height',
    figsize=(15, 8)
)
```

```python
# Для 33 этажей нет данных, уберем пока это значение. Группируем и находим медианные значения для высоты потолка
# по этажности дома
data_group_ceiling_height = (data.groupby(['floors_total'], as_index=False)['ceiling_height']
                             .median()
                             .set_index('floors_total'))
data_group_ceiling_height = data_group_ceiling_height.dropna()
data_group_ceiling_height
```

Минимальной нормой считается высота потолка 2,4 метра. Согласно этим данным, примем, что все, что менее 2,40 выброс.

```python
# Заменяем выбросы медианным значением с использованием map() и строим диаграмму размаха
data.loc[
    (data.ceiling_height >= 4) 
    | (data.ceiling_height < 2.4)
    | (data.ceiling_height.isna()),
    'ceiling_height'
] = data[
    (data.ceiling_height >= 4) 
    | (data.ceiling_height < 2.4)
    | (data.ceiling_height.isna())
].floors_total.map(
    data_group_ceiling_height['ceiling_height']
)

data.boxplot(
    by='floors_total', 
    column='ceiling_height',
    figsize=(15, 8)
)
```

Для 33 этажного здания отсутствуют данные, а для 37 этажей аномально низкое значение. Заменим эти данные на медианное значение рассчитаное по медианнам для остальной этажности.

```python
# Рассчитываем медианное значение по медианнам всей этажности
data_group_ceiling_height['ceiling_height'].median()
```

```python
# Заменяем высоту потолков в 33 и 37 этажном доме на медианное значение
data.loc[data.floors_total.isin([33,37]),'floors_total'] = 2.7
```

```python
# Оставшиеся значения NaN также заменяем медианным значением и сторим диаграмму размаха
data = data.fillna(value={'ceiling_height': 2.7})

data.boxplot(
    by='floors_total', 
    column='ceiling_height',
    figsize=(15, 8)
)
data.isna().sum()
```

Значения превышающие 1,5IQR, но не превышающие 4 метра не отбрасываем, так как высота потолков в сталинских домах достигает 3–4 м.

```python
# Строим гистограмму для ceiling_height
data.hist('ceiling_height', bins=50)
```

Из гистрограммы видно, что наиболее встречающее значение высоты потолков около 2.5-2.7

#### Заполнение пропусков в столбце living_area

```python
# Находим коэффициент корреляции между значениями total_area и living_area
data['total_area'].corr(data['living_area'])
```

```python
data.plot(x='total_area', y='living_area', kind='scatter', alpha=0.3) 
```

Значение корреляции очень близко к 1. Зависимость практически линейна. Следовательно при увеличении значения total_area увеличивается значение living_area. Найдем медианное значение коэффициента, показывающего во сколько раз отличается living_area и total_area.

```python
# Находим медианное значение
living_area_coeff = (data['living_area'] / data['total_area']).median()
living_area_coeff
```

```python
# Заменим пропуски в living_area
data.loc[data.living_area.isna(), 'living_area'] = (data.loc[data.living_area.isna(),'total_area']
                                                    * living_area_coeff)
```

#### Замена пропусков в столбце locality_name

```python
# Заменим NaN в locality_name на unknown
data['locality_name'] = data['locality_name'].fillna('unknown')
data['locality_name'].isna().sum()
```

#### Замена пропусков в столбце city_centers_nearest

```python
# очистим названия населённых пунктов от указания административных единиц.
# столбец с такими данными будем использовать далее.

STOP_WORDS = ("деревня", "поселок", "городской", "тип", "село")

def clean_locality_name(row):
    name = row.locality_name.replace('ё', 'е')
    name = ' '.join(m.lemmatize(name))
    for stop_word in STOP_WORDS:
        name = name.replace(stop_word, '')
    name = name.replace(' - ', '-')
    name = name.strip()
    return name

data['clean_locality_name'] = data.apply(clean_locality_name, axis=1)

data[['locality_name', 'clean_locality_name']].head(10)
```

```python
#Заполним пропуски в столбце city_centers_nearest группируя по столбцу locality_name 
data['city_centers_nearest'] = (data['city_centers_nearest']
                               .fillna(data.groupby('clean_locality_name')['city_centers_nearest']
                                      .transform('median')))
data.isna().sum()
```

```python
print(data.query('locality_name != "санкт-петербург"').copy().isna().sum())
print(data.query('locality_name == "санкт-петербург"').copy().isna().sum())
```

```python
# Убедимся, что у студий нет площади кухни
data[(data['studio'] == True)][['kitchen_area', 'studio']].isna().sum()
```

```python
# Убедися, что у open_plan нет площади кухни
data[
    (data['open_plan'] == True)][['kitchen_area', 'open_plan']].isna().sum()
```

```python
# У студий и квартир с открытой планировкой нет кухни, следовательно можно заполнить пропуcки 0
data.loc[
    (data.studio == True)
    & (data.open_plan == True),
    'kitchen_area'
] = data[
    (data.studio == True)
    & (data.open_plan == True)
].fillna(0)
#data['kitchen_area'].isna().sum()
```

```python
# Заменяем пропуски в столбце kitchen_area группируя по общей площади квартиры
def total_area_group(row):
    if  0<= row.total_area <= 35:
        return 'A'
    elif 35<row.total_area<=64:
        return 'B'
    elif 64<row.total_area<=80:
        return 'C'
    elif 80<row.total_area<=100:
        return 'D'
    elif row.total_area>100:
        return 'E'
    return
data['total_area_group'] = data.apply(total_area_group, axis=1)
data['kitchen_area'] = (data['kitchen_area']
                              .fillna(data.groupby('total_area_group')['kitchen_area']
                                      .transform('median')))
```

```python
data.hist('kitchen_area',bins=50)
data['kitchen_area'].describe()
```

Вид гистограммы не изменился, после заполнения пропусков медианное и среднее значение уменьшились приблизительно на 0,1.

```python
# Заменем пропуски, которые возможно заменить, с столбцax airports_nearest, parks_nearest, ponds_nearest:
for series in ('airports_nearest', 'parks_nearest', 'ponds_nearest'):
    data[series] = data[series].fillna(
        data.groupby('clean_locality_name')[series].transform('median')
    )
data.isna().sum()
```

```python
test = SPB[['city_centers_nearest_km', 'last_price', 'data_publication']].copy()
test['data_publication'] = pd.DatetimeIndex(test['data_publication']).year
test.head()
for i in test['data_publication'].unique():
    print('year: ', i)
    print(test[test['data_publication'] == i][['city_centers_nearest_km', 'last_price']].corr(), end= '\n\n')
```

Если посмотреть корреляцию по годам, то можно заметить, что она была сильнее в 2014 году. Тогда, когда была выше стоимость и меньше продаж. Возможно застройка спальных районов новостройками сильно изменила рынок недвижимости Питера.
